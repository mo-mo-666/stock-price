```python
import numpy as np
import pandas as pd
import pandas_datareader.data as web
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import bokeh
from bokeh.plotting import figure, output_notebook, show
from ipywidgets import interact

%matplotlib inline
%config InlineBackend.figure_formats = {'png', 'retina'}
output_notebook()
```

```python
start = "1950/1/1"

df = web.DataReader('^N225', 'yahoo', start)
df
```

```python
p1 = figure(width = 1000 , height = 400 ,x_axis_type = 'datetime', title="NIKKEI")
p1.line(df.index, df["Adj Close"], legend_label = 'NIKKEI')
show(p1)
```

- 1986-1991 バブル景気
- 1991-1993 バブル崩壊 (第1次平成不況)
- 1999-2000 ITバブル
- 2000-2002 IT不況 (第3次平成不況)
- 2007 リーマンショック → 世界金融危機 (2007-2010)
- 2012-2020 アベノミクス
- 2020- コロナショック

```python
p2 = figure(width = 1000 , height = 400 ,x_axis_type = 'datetime', y_axis_type="log", title="Log of NIKKEI")
p2.line(df.index, df["Adj Close"], legend_label = 'NIKKEI')
show(p2)
```
